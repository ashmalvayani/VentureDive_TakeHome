**Variance Threshold**
The VarianceThreshold technique is a useful solution for identifying uninformative features in a dataset. By removing features with low variance, it helps simplify the data and improve model performance by reducing noise and irrelevant information. Its computational efficiency and interpretability make it a practical choice for feature selection.

```python
from sklearn.feature_selection import VarianceThreshold

# Separate the features and labels
X = class_updated.drop('Labels', axis=1)
y = class_updated['Labels']

selector = VarianceThreshold()
selector.fit(X)

selected_features = selector.get_support()

uninformative_features = X.columns[~selected_features]

print("Uninformative Features:")
print(uninformative_features)
```

**1 C) Redundant Features**

*To find redundant features in a dataset, one approach is to calculate the correlation matrix among all the features and identify highly correlated pairs.*

```python
df = class_updated

corr_matrix = df.corr().abs()

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

threshold = 0.8
redundant_features = {}

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if corr_matrix.iloc[i, j] >= threshold:
            feature_i = corr_matrix.columns[i]
            feature_j = corr_matrix.columns[j]

            if feature_i in redundant_features.keys():
              list_ = redundant_features[feature_i]
              list_.append(feature_j)
              redundant_features[feature_i] = list_
            else:
              list_ = [feature_j]
              redundant_features[feature_i] = list_

print("Redundant Feature Pairs:")
for feature in redundant_features.keys():
  print(feature, " : ",  redundant_features[feature])
```

**Conclusion:** The code generates a correlation matrix heatmap using Seaborn, where highly correlated features will have a higher correlation coefficient value. Redundant features are identified by setting a correlation threshold as 0.8, and any feature pairs with correlation coefficients above the threshold are considered redundant. The redundant feature pairs are then printed for further analysis.

# **3. Train a classifier to predict the labels from the provided features, assuming that each type  of misclassification error is equally costly. Clearly label each step of the process wrt your  findings from 1 above.**

Step1: Import the relevant libraries

```python
from imblearn.over_sampling import SMOTE
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

Step2: since we know that our data is imbalanced, let's apply SMOTE to balance the dataset.

```python
X = informative_features_df.drop('Labels', axis=1)
y = informative_features_df['Labels']

smote = SMOTE()
x_smote, y_smote = smote.fit_resample(X,y)
```

Step3: We have ensured that we only select the informative features from the dataset and not the uninformative features.

```python
X_train, X_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size=0.2, random_state=42)
```

Step4: Initialize the classifers and train them.

```python
# Initializing the models
logreg = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
xgb_classifier = xgb.XGBClassifier()

# Training the models
logreg.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
xgb_classifier.fit(X_train, y_train)


# Making predictions
logreg_pred = logreg.predict(X_test)
dt_pred = dt.predict(X_test)
rf_pred = rf.predict(X_test)
xgb_pred = xgb_classifier.predict(X_test)

# Calculating accuracy
logreg_accuracy = accuracy_score(y_test, logreg_pred)
dt_accuracy = accuracy_score(y_test, dt_pred)
rf_accuracy = accuracy_score(y_test, rf_pred)
xgb_accuracy = accuracy_score(y_test, xgb_pred)

# Printing the accuracy scores
print("Logistic Regression Accuracy:", logreg_accuracy)
print("Decision Tree Accuracy:", dt_accuracy)
print("Random Forest Accuracy:", rf_accuracy)
print("XGBoost Accuracy:", xgb_accuracy)
```

```python
classifiers = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost']
reports = [classification_report(y_test, logreg_pred, output_dict=True),
           classification_report(y_test, dt_pred, output_dict=True),
           classification_report(y_test, rf_pred, output_dict=True),
           classification_report(y_test, xgb_pred, output_dict=True)]

# Configure subplots
fig, axes = plt.subplots(nrows=len(classifiers), figsize=(8, 4 * len(classifiers)))

# Generate plots
for ax, classifier, report in zip(axes, classifiers, reports):
    df_report = pd.DataFrame(report).iloc[:-1, :].T
    cmap = sns.color_palette("coolwarm", as_cmap=True)
    sns.heatmap(df_report, annot=True, cmap=cmap, ax=ax)
    ax.set_title(f"Classification Report - {classifier}")

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

# **4. Estimate the expected predictive accuracy of the trained classifier using an appropriate  metric.**

```python
# from sklearn.metrics import accuracy_score
# import pandas as pd
# import seaborn as sns
# import matplotlib.pyplot as plt


accuracies = [logreg_accuracy, dt_accuracy, rf_accuracy, xgb_accuracy]
classifiers = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost']
data = {'Classifier': classifiers, 'Accuracy': accuracies}
df_accuracy = pd.DataFrame(data)

# Plot the accuracy scores
plt.figure(figsize=(6, 3))
sns.barplot(x='Accuracy', y='Classifier', data=df_accuracy, palette='Blues')
plt.title('Classifier Accuracy')
plt.xlabel('Accuracy')
plt.ylabel('Classifier')

for i, acc in enumerate(accuracies):
    plt.text(acc, i, f'{acc:.2f}', ha='left', va='center', color='white')

# Show the plot
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, f1_score

classifiers = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost']
models = [logreg, dt, rf, xgb_classifier]

auc_roc_scores = []
f1_scores = []
for model in models:
    y_pred_prob = model.predict_proba(X_test)
    # Convert y_pred_prob to one-vs-rest format
    y_pred_ovr = np.zeros_like(y_pred_prob)
    y_pred_ovr[np.arange(len(y_pred_prob)), y_pred_prob.argmax(axis=1)] = 1
    auc_roc_scores.append(roc_auc_score(y_test, y_pred_ovr, multi_class='ovr'))
    f1_scores.append(f1_score(y_test, y_pred_ovr.argmax(axis=1), average='macro'))

data = {'Classifier': classifiers, 'AUC-ROC': auc_roc_scores, 'F1 Score': f1_scores}
df_metrics = pd.DataFrame(data)

plt.figure(figsize=(9,3))
sns.barplot(x='AUC-ROC', y='Classifier', data=df_metrics, palette='Oranges', label='AUC-ROC')
sns.barplot(x='F1 Score', y='Classifier', data=df_metrics, palette='Blues', label='F1 Score')
plt.title('Classifier Evaluation Metrics')
plt.xlabel('Score')
plt.ylabel('Classifier')
plt.legend()

# Show the plot
plt.show()
```

**Conclusion:**

AUC-ROC (Area Under the Receiver Operating Characteristic Curve) and F1 score are evaluation metrics that are robust to class imbalance.
*   It is unaffected by class imbalance because it assesses the model's performance by plotting the true positive rate against the false positive rate.
*   F1 score provides a balanced measure of the model's accuracy, taking into account both the positive and negative classes.
